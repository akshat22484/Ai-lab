```python
from queue import PriorityQueue


graph = {
    'a': {'b': 4, 'c': 3},
    'b': {'a': 4, 'f': 5, 'e': 12},
    'c': {'a': 3, 'd': 7, 'e': 10},
    'd': {'c': 7, 'e': 2},
    'e': {'c': 10, 'd': 2, 'b': 12, 'z': 5},
    'f': {'b': 5, 'z': 16},
    'z': {}
}

h = {
    'a': 14,
    'b': 12,
    'c': 11,
    'd': 6,
    'e': 4,
    'f': 11,
    'z': 0
}

def a_star(start, goal):
    pq = PriorityQueue()
    pq.put((h[start], 0, start, [start]))
    visited = set()

    step_count = 0
    print(f"Initial PQ: [(f={h[start]}, g={0}, node='{start}', path={[start]})]")

    while not pq.empty():
        step_count += 1
        f, g, node, path = pq.get()

        print(f"\n--- Step {step_count} ---")
        print(f"Current node: '{node}' (f={f}, g={g})")
        print(f"Current path: {' -> '.join(path)}")

        if node in visited:
            print(f"Node '{node}' already visited. Skipping.")
            continue
        visited.add(node)
        print(f"Adding '{node}' to visited. Visited nodes: {visited}")

        if node == goal:
            print(f"Goal '{goal}' reached!")
            print(f"Optimal Path: {' -> '.join(path)}")
            print(f"Total Cost (g): {g}")
            return

        print(f"Exploring neighbors of '{node}':")
        for neighbor, cost in graph[node].items():
            if neighbor not in visited:
                g_new = g + cost
                f_new = g_new + h[neighbor]
                pq.put((f_new, g_new, neighbor, path + [neighbor]))
                print(f"  - Neighbor '{neighbor}': cost={cost}, g_new={g_new}, h_neighbor={h[neighbor]}, f_new={f_new}. Added to PQ.")
            else:
                print(f"  - Neighbor '{neighbor}' already visited. Skipping.")


        temp_list = []
        while not pq.empty():
            temp_list.append(pq.get())
        temp_list.sort()
        for item in temp_list:
            pq.put(item)

        pq_display = [(f_val, g_val, n, p) for f_val, g_val, n, p in temp_list]
        print(f"Priority Queue state: {pq_display}")

    print("No path found.")

a_star('a', 'z')
```
